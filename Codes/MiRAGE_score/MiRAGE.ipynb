## MiRAGE: 
Mining Relationships for Advanced Generative Evaluation in Drug Repositioning

```python
import pandas as pd
from tqdm import tqdm
```

```python
#Building up an empty dataframe for MiRAGE scores for each drug_disease pair 
#Note: Obviously the number of similarities(j) could be different in this case j=2 and total featueres will be 6(2j+2)
df_scores = pd.DataFrame(columns=['drugID', 'diseaseID', 
                                  'p_score','q_score','count_drug','count_disease'])
df_scores
```

```python
#Using a disease-disease similarity as an example to representing the structure we worked with
disease_sim = pd.read_csv("dataset/disease_sim.csv")
disease_sim
```

```python
#Using a drug-drug similarity as an example to representing the structure we worked with
drug_sim = pd.read_csv("dataset/drug_sim.csv")
drug_sim
```

```python
mapping = pd.read_csv("dataset/mapping.csv")
mapping80 = pd.read_csv("dataset/mapping80.csv")
mapping80
```

```python
#All drugs and all diseases 
all_drugs = mapping['DrugID'].unique().tolist()
all_diseases = mapping['DiseaseID'].unique().tolist()
```

```python
for drug_test in tqdm(all_drugs):
    for disease_test in all_diseases:
        
        #Extract As and Bd From train mapping
        disease_list_one = mapping80[mapping80['drug_ID'] == drug_test]['disease_ID'].tolist()
        drug_list_one = mapping80[mapping80['disease_ID'] == disease_test]['drug_ID'].tolist()
        
        # We diligently calculate scores even for single instances within our training mappings.
        disease_list_one = [item for item in disease_list_one if item != disease_test]
        drug_list_one = [item for item in drug_list_one if item != drug_test]
        
        #Set default values to zero
        q1_score = 0
        p1_score = 0
        
        # Compute the maximum similarity as MiRAGE score
        if drug_list_one:
            p1_score = max(max(drug_sim[drug_sim['Unnamed: 0'] == drug_test][drug_list_one].values.tolist()))

        if disease_list_one:
            q1_score = max(disease_sim[disease_sim['DiseaseID'] == disease_test][disease_list_one].values.flatten())


        #Adding len of As and Bd and use them as two features
        count_drug = len(drug_list_one)
        count_disease = len(disease_list_one)
        
        #Filling the df 
        example_data = [drug_test, disease_test, p1_score, q1_score, count_drug, count_disease]  


        # Add example data to the last row
        df_scores.loc[len(df_scores)] = example_data

#Adjusted MiRAGE 
df_scores['adj_p'] = df_scores['p_score'] * df_scores['count_disease']
df_scores['adj_q'] = df_scores['q_score'] * df_scores['count_drug']
```
